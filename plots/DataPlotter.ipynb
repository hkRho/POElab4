```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def data_plot(filename, startval, endval):
    data = np.genfromtxt('LFSensorData.csv', delimiter=',')
    time=data[startval:endval,0]/1000
    lIR=data[startval:endval,1]*5.0/1023
    lTV=(np.ones(lIR.shape)*900)*5.0/1023
    rIR=data[startval:endval,2]*5.0/1023
    rTV=(np.ones(rIR.shape)*870)*5.0/1023
    lMS=data[startval:endval,3]
    rMS=data[startval:endval,4]
    fig1 = plt.figure(figsize=(20,10))
    plt.plot(time, lIR,
             time, lTV,
             time, rIR,
             time, rTV,
             time, lMS/12,
             time, rMS/12)
    plt.xlabel('Time (s)')
    plt.ylabel('Voltage (V) / Motor Speed Command (unitless)')
    plt.title('IR sensor data / Motor Speed Command over time')
    plt.legend(['Left IR Sensor',
                'Left Threshold Value',
                'Right IR Sensor',
                'Right Threshold Value',
                'Left Motor Speed',
                'Right Motor Speed'])
    plt.savefig("FullGraph.png", dpi='figure')
    fig2 = plt.figure(figsize=(20,10))
    plt.plot(time, lIR,
             time, lTV,
             time, lMS/12)
    plt.xlabel('Time (s)')
    plt.ylabel('Voltage (V) / Motor Speed Command (unitless)')
    plt.title('IR sensor data / Motor Speed Command over time')
    plt.legend(['Left IR Sensor',
                'Left Threshold Value',
                'Left Motor Speed'])
    plt.savefig("LeftSideGraph.png", dpi='figure')
    fig3 = plt.figure(figsize=(20,10))
    plt.plot(time, rIR,
             time, rTV,
             time, rMS/12)
    plt.xlabel('Time (s)')
    plt.ylabel('Voltage (V) / Motor Speed Command (unitless)')
    plt.title('IR sensor data / Motor Speed Command over time')
    plt.legend(['Right IR Sensor',
                'Right Threshold Value',
                'Right Motor Speed'])
    plt.savefig("RightSideGraph.png", dpi='figure')
    
    plt.show()
```

```python
data_plot('LFSensorData.csv',0,2098)
```
